# Objective
In this notebook, I retrieve the data from the aqi API, load it as two seperate json files (since the rules state that you can only grab 5 parameters at a time), and lastly, convert the files to data frames, which I merge, and lastly, convert to the CSV file "missouri_air_quality.csv."

```python
import pandas as pd
import json
import requests
```

```python
url = "https://aqs.epa.gov/data/api/list/parametersByClass?email=ltfyle01@gmail.com&key=orangebird59&pc=CRITERIA"
response = requests.get(url)
data = json.loads(response.text)
data.keys()
```

```python
param_dict=data["Data"]
```

```python
param_dict
```

```python
url = "https://aqs.epa.gov/data/api/dailyData/byState?&email=ltfyle01@gmail.com&key=orangebird59&param=14129,42101,42401,42602,44201&bdate=20240101&edate=20241231&state=29"
state_data1 = requests.get(url)
state_data1 = state_data1.json()
```

```python
url = "https://aqs.epa.gov/data/api/dailyData/byState?&email=ltfyle01@gmail.com&key=orangebird59&param=81102,85129,88101&bdate=20240101&edate=20241231&state=29"
state_data2 = requests.get(url)
state_data2 = state_data2.json()
```

```python
with open("data_1.json", "w") as f:
    json.dump(state_data1, f)

with open("data_2.json", "w") as f:
    json.dump(state_data2, f)
```

```python
data1 = json.load(open("data_1.json", "r"))
data2 = json.load(open("data_2.json", "r"))
```

```python
df1 = pd.DataFrame(data1['Data'])
df2 = pd.DataFrame(data2['Data'])
```

```python
df_combined = pd.concat([df1, df2], ignore_index=True)
```

```python
df_combined['parameter'].unique()
```

```python
df_combined.to_csv("missouri_air_quality_data.csv", index=False)
```
