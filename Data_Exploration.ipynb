```python
def summarize_data(myData):

    'Returns a description of each column'
    
    dtypes=myData.dtypes
    counts=myData.apply(lambda col: col.count())
    nulls=myData.apply(lambda col: col.isnull().sum())
    nan_ratio=nulls/counts
    uniques=myData.apply(lambda col: col.unique())
    n_uniques=myData.apply(lambda col: col.nunique())

    
    maxs = myData.apply(lambda col: col.max() if col.dtype in ['int64','float64'] else '...')
    mins = myData.apply(lambda col: col.min() if col.dtype in ['int64','float64'] else '...')

    cols = {'dtypes':dtypes,'counts':counts,'nulls':nulls,'nan_ratio':nan_ratio,'uniques':uniques,'n_uniques':n_uniques,'maxs':maxs,'mins':mins}

    return pd.DataFrame(data=cols)
```

```python
data_summary=summarize_data(df)
data_summary
```
